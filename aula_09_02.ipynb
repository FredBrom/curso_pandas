```python
import pandas as pd

comb_df = pd.read_excel('ca-2021-02.xlsx')
```

## Inserir uma coluna 'Obs' ao data frame e, se a cidade for a sua cidade, coloque 'Melhor cidade', caso contrario, uma string em branco ' '.

```python
#Usando List Comprehension:
comb_df['Obs'] = ['Melhor Cidade' if municipio == 'BELO HORIZONTE' else ' ' for municipio in comb_df['Municipio']]
```

```python
display(comb_df[['Municipio','Obs']].loc[comb_df['Municipio'].isin(['BELO HORIZONTE', 'BETIM'])])
```

## Preencher a coluna 'Status' com "CARO" caso o valor de venda esteja maior que a média e "BARATO" caso o valor de venda esteja menor ou igual a média

```python
preco_medio = comb_df['Valor de Venda'].mean()

comb_df['Status'] = ['BARATO' if valor <= preco_medio else 'CARO' for valor in comb_df['Valor de Venda']]

display(comb_df[['Revenda', 'Valor de Venda', 'Status']].head())
```

Calcular postos de gasolina por habitantes temos na amostragem de combustíveis 

```python
habitantes_df = pd.read_csv('ibge_num_habitantes_estimado.csv', delimiter=';')

#renomear a coluna estado para 'Estado - Sigla'

habitantes_df.rename(columns={'Estado':'Estado - Sigla'}, inplace=True)
display(habitantes_df)
```

```python
#fazer um MERGE dos dois data-frames

comb_df = comb_df.merge(habitantes_df, how='inner', on=['Municipio', 'Estado - Sigla'])

comb_df.info()
```

```python
#Apaga a coluna que estava vazia
comb_df.dropna(axis='columns')
comb_df.info()
```

```python
comb_df.drop(labels=['Regiao - Sigla', 'Nome da Rua', 'Cep', 'Data da Coleta', 'Valor de Venda', 'Unidade de Medida', 'Bandeira', 'Obs', 'Status', 'Numero Rua', 'Complemento', 'Bairro', 'Valor de Compra'],axis=1, inplace=True)
comb_df.info()
```

```python
comb_df.drop(labels='Produto', axis=1, inplace=True)
comb_df.drop_duplicates(inplace=True)
comb_df.info()
```

```python
#Agrupar e contar quantos postos por cidade:
postos_por_municipio_df = comb_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={'Revenda':'Numero de Postos'}, inplace=True)
postos_por_municipio_df.info()
```
