## Criando um novo data-frame

```python
ca_df = combustiveis_df[['Revenda', 'Municipio', 'Produto','Valor de Venda']]

display(ca_df)
```

```python
#Exibe a 4 linha
print(ca_df.loc[3])
```

```python
for i in range(ca_df.shape[0]):
  if ca_df.loc[i]["Municipio"] == "BELO HORIZONTE":
    print(ca_df.loc[i])
```

```python
#Filtrar por um intervalo especifico
display(ca_df.loc[9:19])
```

```python
from pandas.io.formats.style_render import com
#Criar um data-frame gas_df contendo apenas as 4 colunas (Revenda, Municipio, Produto, Valor de Venda)
# somente com combustivel sendo gasolina comum e exibir na tela

gas_df = ca_df.loc[ca_df['Produto'] == 'GASOLINA']

display(gas_df)
```

```python
# Encontrar o posto com a gasolina mais cara do pais nesse mês.
display(gas_df[['Revenda', 'Municipio','Valor de Venda']].max())
```
